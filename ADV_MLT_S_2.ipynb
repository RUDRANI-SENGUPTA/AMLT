```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
wine=pd.read_csv('https://gist.githubusercontent.com/tijptjik/9408623/raw/b237fa5848349a14a14e5d4107dc7897c21951f5/wine.csv')
```

```python
wine
```

```python
X=wine.drop(['Wine'],axis=1)
X
```

```python
X.shape
```

```python
y=wine['Wine']
```

```python
y
```

```python
y.unique()
```

```python
#Standardisation of features
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_scaled=sc.fit_transform(X)
X_scaled
```

```python
#Building LDA model
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda=LinearDiscriminantAnalysis(n_components=2)
```

```python
type(lda)
```

```python
#Transforming the data
X_lda=lda.fit_transform(X_scaled,y)
```

```python
X_lda
```

```python
X_lda.shape
```

```python
#Visualisation of the data
plt.scatter(X_lda[:,0],X_lda[:,1],c=y);
```

```python
#Locally linear embedding
from sklearn.datasets import make_swiss_roll
```

```python
X,y=make_swiss_roll(n_samples=1000,random_state=100)
X
```

```python
y
```

```python
#Building the model
from sklearn.manifold import LocallyLinearEmbedding
```

```python
lle=LocallyLinearEmbedding(n_neighbors=10,random_state=100)
```

```python
#Transforming the data
X_lle=lle.fit_transform(X)
```

```python
X_lle
```

```python
#Visualisation
plt.scatter(X_lle[:,0],X_lle[:,1],c=y)
```
