```python
#Create a random sample of 1K from 70K
np.random.seed(100)
sample=np.random.choice(X.shape[0],1000)
print(sample)
```

```python
#Creating a subset of 1K samples of X,y
X1=X.iloc[sample,:]
X1.shape
```

```python
X1.head()
```

```python
y1=y[sample]
y1.shape
```

```python
y1.head()
```

```python
#Building the TSNE model
from sklearn.manifold import TSNE#components is the dimension onto which it is transformed
tsne=TSNE(n_components=2,perplexity=30)
```

```python
X_tsne=tsne.fit_transform(X1)
X_tsne.shape
```

```python
X_tsne[0]
```

```python
#Plotting transformed data points
plt.scatter(X_tsne[:,0],X_tsne[:,1]);
```

```python
plt.scatter(X_tsne[:,0],X_tsne[:,1],c=y1.astype(float));
```

```python
#Creating a dataframe
X_df=pd.DataFrame({'X0':X_tsne[:,0],'X1':X_tsne[:,1],'Label':y1})
X_df
```

```python
plt.figure(figsize=(15,12))
sns.lmplot(data=X_df,x='X0',y='X1',hue='Label')
```

```python
plt.figure(figsize=(15,12))
sns.lmplot(data=X_df,x='X0',y='X1',hue='Label',fit_reg=False)#data points are the numbers, each of which has 784 dimensions
```
